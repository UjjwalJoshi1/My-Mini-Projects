**Online news article data- I**

```python
import pandas as AAD

ON_NeWs_ATL_I = AAD.read_csv('BBC News Train.csv')
ON_NeWs_ATL_I
```

```python
ON_NeWs_ATL_I['Category'].value_counts()
```

* Five classes in output category.....

** Pre-process

```python
del ON_NeWs_ATL_I['ArticleId']
```

```python
ON_NeWs_ATL_I.isnull().values.sum()
```

```python
print("rows shape - duplicates found")
(ON_NeWs_ATL_I[ON_NeWs_ATL_I.duplicated()]).shape[0]
```

```python
ON_NeWs_ATL_I =ON_NeWs_ATL_I.drop_duplicates()
ON_NeWs_ATL_I
```

```python
## """NLP process"""
import nltk as LLK
LLK.download('stopwords')

import regex as GGG
from nltk.corpus import stopwords as STT
STT_L = STT.words('english')

from nltk.stem.porter import PorterStemmer as PSS
PSS_L = PSS()

def Research(ONA):
    ONA = ONA.lower()
    ONA = GGG.sub("[^a-z]+", " ", ONA)
    ONA = " ".join([A for A in ONA.split() if A not in STT_L])
    ONA=' '.join([PSS_L.stem(A) for A in ONA.split()])
    return ONA

EMpty_L=[]
for c in ON_NeWs_ATL_I['Text']:
    EMpty_L.append(Research(c))
ON_NeWs_ATL_I['Text']=EMpty_L
ON_NeWs_ATL_I
```

```python
import seaborn as BBS
import matplotlib.pyplot as YYT
YYT.figure(figsize=(7, 5))

BBS.countplot(y ='Category', data = ON_NeWs_ATL_I, color='maroon')
YYT.title("Online News Article Category")
YYT.show()
```

* Sport category having high count of news.

```python
ON_NeWs_ATL_I.to_csv('ON_NeWs_ATL_I.csv', index=False)
ON_NeWs_ATL_I.shape
```
